```python
import pandas as pd

# Sample DataFrame with buy/sell status
data = {
    "Stock": ["ASHOKLEY", "TATAMOTORS", "TATAMOTORS", "ASHOKLEY", "ASHOKLEY", "ASHOKLEY", "ASHOKLEY", "ASHOKLEY", "ASHOKLEY"],
    "Status": ["BUY", "BUY", "SELL", "SELL", "BUY", "SELL", "SELL", "BUY", "BUY"],
    "Qty.": [1000, 250, 250, 1000, 2000, 2000, 2000, 1000, 1000],
    "Avg. Price": [125.70, 490.55, 492.10, 125.96, 125.70, 125.95, 125.95, 125.60, 125.65]
}

df = pd.DataFrame(data)

# Define turnover charge rates
charge_rates = {
    "A_B_non-exclusive": 375,  # ₹375 per crore (default)
    "M_MT_TS_MS": 275,
    "SS_ST": 100000,
    "X_XD_XT_Z_ZP": 10000
}

# Function to calculate turnover charge
def calculate_turnover_charge(row, group="A_B_non-exclusive"):  
    turnover = row["Qty."] * row["Avg. Price"]
    charge_rate = charge_rates.get(group, 375)  
    return (charge_rate / 10**7) * turnover

# Function to calculate brokerage
def calculate_brokerage(row):
    brokerage = row["Qty."] * row["Avg. Price"] * 0.0003  
    return min(brokerage, 20)

# Function to calculate STT/CTT (Only on Sell Trades)
def calculate_stt_ctt(row):
    if row["Status"] == "SELL":
        return row["Qty."] * row["Avg. Price"] * 0.00025  # 0.025%
    return 0

# Function to calculate SEBI charges (₹10 per crore)
def calculate_sebi_charges(row):
    turnover = row["Qty."] * row["Avg. Price"]
    return (10 / 10**7) * turnover

# Function to calculate Stamp Duty (Only on Buy Trades, ₹300 per crore or 0.003%)
def calculate_stamp_duty(row):
    if row["Status"] == "BUY":
        turnover = row["Qty."] * row["Avg. Price"]
        return (300 / 10**7) * turnover  # ₹300 per crore (0.003%)
    return 0

# Apply calculations
df["Turnover"] = df["Qty."] * df["Avg. Price"]
df["Turnover Charge"] = df.apply(lambda row: calculate_turnover_charge(row), axis=1)
df["Brokerage"] = df.apply(lambda row: calculate_brokerage(row), axis=1)
df["STT/CTT"] = df.apply(lambda row: calculate_stt_ctt(row), axis=1)
df["SEBI Charges"] = df.apply(lambda row: calculate_sebi_charges(row), axis=1)
df["Stamp Duty"] = df.apply(lambda row: calculate_stamp_duty(row), axis=1)

# Calculate Total Charges
df["Total Charges"] = df["Turnover Charge"] + df["Brokerage"] + df["STT/CTT"] + df["SEBI Charges"] + df["Stamp Duty"]

# Display the DataFrame
print(df)

# Print Total Charges
total_turnover = df["Turnover"].sum()
total_turnover_charge = df["Turnover Charge"].sum()
total_brokerage = df["Brokerage"].sum()
total_stt_ctt = df["STT/CTT"].sum()
total_sebi_charges = df["SEBI Charges"].sum()
total_stamp_duty = df["Stamp Duty"].sum()
total_charges = df["Total Charges"].sum()

print(f"\nTotal Turnover: ₹{total_turnover:,.2f}")
print(f"Total Turnover Charge: ₹{total_turnover_charge:,.2f}")
print(f"Total Brokerage: ₹{total_brokerage:,.2f}")
print(f"Total STT/CTT: ₹{total_stt_ctt:,.2f}")
print(f"Total SEBI Charges: ₹{total_sebi_charges:,.2f}")
print(f"Total Stamp Duty: ₹{total_stamp_duty:,.2f}")
print(f"Total Charges: ₹{total_charges:,.2f}")
```

```python
import pandas as pd

# Sample DataFrame with stock transactions
data = {
    "Instrument": ["ASHOKLEY", "TATAMOTORS", "TATAMOTORS", "ASHOKLEY", "ASHOKLEY", "ASHOKLEY", "ASHOKLEY", "ASHOKLEY", "ASHOKLEY"],
    "Type": ["BUY", "BUY", "SELL", "SELL", "BUY", "SELL", "SELL", "BUY", "BUY"],
    "Qty.": [1000, 250, 250, 1000, 2000, 2000, 2000, 1000, 1000],
    "Avg. Price": [125.70, 490.55, 492.10, 125.96, 125.70, 125.95, 125.95, 125.60, 125.65]
}

df = pd.DataFrame(data)

# Define turnover charge rates
charge_rates = {
    "A_B_non-exclusive": 375,  # ₹375 per crore (default)
    "M_MT_TS_MS": 275,
    "SS_ST": 100000,
    "X_XD_XT_Z_ZP": 10000
}

# Function to calculate weighted average price
def weighted_avg_price(group):
    return (group["Qty."] * group["Avg. Price"]).sum() / group["Qty."].sum()

# Function to calculate turnover charge
def calculate_turnover_charge(row, group="A_B_non-exclusive"):  
    turnover = row["Qty."] * row["Avg. Price"]
    charge_rate = charge_rates.get(group, 375)  
    return (charge_rate / 10**7) * turnover

# Function to calculate brokerage
def calculate_brokerage(row):
    brokerage = row["Qty."] * row["Avg. Price"] * 0.0003  
    return min(brokerage, 20)

# Function to calculate STT/CTT (Only on Sell Trades)
def calculate_stt_ctt(row):
    if row["Type"] == "SELL":
        return row["Qty."] * row["Avg. Price"] * 0.00025  # 0.025%
    return 0

# Function to calculate SEBI charges (₹10 per crore)
def calculate_sebi_charges(row):
    turnover = row["Qty."] * row["Avg. Price"]
    return (10 / 10**7) * turnover

# Function to calculate GST (18% on Brokerage)
def calculate_gst(row):
    return 0.18 * row["Brokerage"]

# Function to calculate Stamp Duty (Only on Buy Trades, ₹300 per crore or 0.003%)
def calculate_stamp_duty(row):
    if row["Type"] == "BUY":
        turnover = row["Qty."] * row["Avg. Price"]
        return (300 / 10**7) * turnover  # ₹300 per crore (0.003%)
    return 0

# Apply calculations
df["Turnover"] = df["Qty."] * df["Avg. Price"]
df["Turnover Charge"] = df.apply(lambda row: calculate_turnover_charge(row), axis=1)
df["Brokerage"] = df.apply(lambda row: calculate_brokerage(row), axis=1)
df["STT/CTT"] = df.apply(lambda row: calculate_stt_ctt(row), axis=1)
df["SEBI Charges"] = df.apply(lambda row: calculate_sebi_charges(row), axis=1)
df["GST"] = df.apply(lambda row: calculate_gst(row), axis=1)
df["Stamp Duty"] = df.apply(lambda row: calculate_stamp_duty(row), axis=1)

# Calculate Total Charges
df["Total Charges"] = df["Turnover Charge"] + df["Brokerage"] + df["STT/CTT"] + df["SEBI Charges"] + df["GST"] + df["Stamp Duty"]

# Group by Stock and Type to get the summary
df_summary = df.groupby(["Instrument", "Type"]).agg({
    "Qty.": "sum",
    "Avg. Price": weighted_avg_price,
    "Turnover": "sum",
    "Brokerage": "sum",
    "STT/CTT": "sum",
    "SEBI Charges": "sum",
    "GST": "sum",
    "Stamp Duty": "sum",
    "Total Charges": "sum"
}).reset_index()

# Rename columns
df_summary.rename(columns={"Avg. Price": "Weighted Avg. Price"}, inplace=True)

# Display Final Analysis
print(df_summary)

# Print Grand Totals
print(f"\nGrand Total Turnover: ₹{df_summary['Turnover'].sum():,.2f}")
print(f"Grand Total Brokerage: ₹{df_summary['Brokerage'].sum():,.2f}")
print(f"Grand Total STT/CTT: ₹{df_summary['STT/CTT'].sum():,.2f}")
print(f"Grand Total SEBI Charges: ₹{df_summary['SEBI Charges'].sum():,.2f}")
print(f"Grand Total GST: ₹{df_summary['GST'].sum():,.2f}")
print(f"Grand Total Stamp Duty: ₹{df_summary['Stamp Duty'].sum():,.2f}")
print(f"Grand Total Charges: ₹{df_summary['Total Charges'].sum():,.2f}")
```

```python
import pandas as pd

# Sample Data (Replace with your actual DataFrame)
data = {
    "Time": ["16/12/21 15:08", "16/12/21 14:13", "16/12/21 13:21"],
    "Type": ["BUY", "BUY", "SELL"],
    "Stock": ["ASHOKLEY", "TATAMOTORS", "TATAMOTORS"],
    "Product": ["MIS", "MIS", "MIS"],
    "Qty.": ["1000/1000", "250/250", "250/250"],
    "Avg. price": [125.70, 490.55, 492.10],
    "Status": ["COMPLETE", "COMPLETE", "COMPLETE"]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# ✅ Clean and Process Quantity Column
df["Qty."] = df["Qty."].str.split("/").str[0]  # Extract first value before '/'
df["Qty."] = pd.to_numeric(df["Qty."], errors="coerce")  # Convert to numeric

# ✅ Calculate Turnover (Price * Quantity)
df["Turnover"] = df["Qty."] * df["Avg. price"]

# ✅ Brokerage Calculation (0.03% or ₹20, whichever is lower)
df["Brokerage"] = df.apply(lambda row: min(row["Turnover"] * 0.0003, 20), axis=1)

# ✅ STT/CTT Charge (0.025% on SELL side only)
df["STT/CTT"] = df.apply(lambda row: row["Turnover"] * 0.00025 if row["Type"] == "SELL" else 0, axis=1)

# ✅ SEBI Charges (₹10 per crore)
df["SEBI"] = df["Turnover"] * (10 / 10**7)  # 10 per crore = 10 / 10^7

# ✅ GST (18% on Brokerage)
df["GST"] = df["Brokerage"] * 0.18  

# ✅ Stamp Charges (0.003% or ₹300 per crore on BUY side only)
df["Stamp Charges"] = df.apply(lambda row: min(row["Turnover"] * 0.00003, 300) if row["Type"] == "BUY" else 0, axis=1)

# ✅ Total Charges
df["Total Charges"] = df["Brokerage"] + df["STT/CTT"] + df["SEBI"] + df["GST"] + df["Stamp Charges"]

# ✅ Stock-wise and Type-wise Analysis with Weighted Avg. Price
grouped = df.groupby(["Stock", "Type"]).agg(
    Total_Qty=("Qty.", "sum"),
    Weighted_Avg_Price=("Avg. price", lambda x: (x * df.loc[x.index, "Qty."]).sum() / df.loc[x.index, "Qty."].sum()),
    Total_Turnover=("Turnover", "sum"),
    Total_Brokerage=("Brokerage", "sum"),
    Total_STT_CTT=("STT/CTT", "sum"),
    Total_SEBI=("SEBI", "sum"),
    Total_GST=("GST", "sum"),
    Total_Stamp_Charges=("Stamp Charges", "sum"),
    Grand_Total_Charges=("Total Charges", "sum")
).reset_index()

# ✅ Display Data
print(df)
print("\nStock-wise & Type-wise Summary:\n", grouped)
```

```python
import pandas as pd

# Sample Data (Replace with actual data)
data = {
    "Type": ["BUY", "BUY", "SELL", "SELL"],
    "Stock": ["ASHOKLEY", "TATAMOTORS", "TATAMOTORS", "ASHOKLEY"],
    "Product": ["MIS", "MIS", "MIS", "MIS"],
    "Qty.": [1000, 250, 250, 1000],
    "Avg. price": [125.70, 490.55, 492.10, 125.96],
    "Status": ["COMPLETE", "COMPLETE", "COMPLETE", "COMPLETE"]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Calculate Turnover
df["Turnover"] = df["Qty."] * df["Avg. price"]

# Calculate Brokerage (0.03% or ₹20 max)
df["Brokerage"] = df["Turnover"].apply(lambda x: min(x * 0.0003, 20))

# Calculate STT/CTT (0.025% on SELL only)
df["STT/CTT"] = df.apply(lambda row: row["Turnover"] * 0.00025 if row["Type"] == "SELL" else 0, axis=1)

# SEBI Charges (₹10 per crore)
df["SEBI"] = df["Turnover"] * (10 / 10**7)

# GST (18% on Brokerage)
df["GST"] = df["Brokerage"] * 0.18

# Stamp Charges (0.003% or ₹300 max on BUY only)
df["Stamp Charges"] = df.apply(lambda row: min(row["Turnover"] * 0.00003, 300) if row["Type"] == "BUY" else 0, axis=1)

# Total Charges
df["Total Charges"] = df[["Brokerage", "STT/CTT", "SEBI", "GST", "Stamp Charges"]].sum(axis=1)

# Stock-wise and Type-wise analysis
grouped = df.groupby(["Stock", "Type"]).agg(
    Qty=("Qty.", "sum"),
    Avg_Price=("Avg. price", lambda x: (x * df.loc[x.index, "Qty."].sum()) / df.loc[x.index, "Qty."].sum()),
    Turnover=("Turnover", "sum"),
    Brokerage=("Brokerage", "sum"),
    STT_CTT=("STT/CTT", "sum"),
    SEBI=("SEBI", "sum"),
    GST=("GST", "sum"),
    Stamp_Charges=("Stamp Charges", "sum"),
    Total_Charges=("Total Charges", "sum")
).reset_index()

# Overall Summary
overall = df.groupby("Stock").agg(
    Gross_PnL=("Turnover", "sum"),
    Total_Charges=("Total Charges", "sum")
).reset_index()
overall["Net_PnL"] = overall["Gross_PnL"] - overall["Total_Charges"]
overall["% Charges on Gross PnL"] = (overall["Total_Charges"] / overall["Gross_PnL"]) * 100

# Save to Excel
with pd.ExcelWriter("Stock_Charges_Summary.xlsx") as writer:
    df.to_excel(writer, sheet_name="Individual Trade Charges", index=False)
    grouped.to_excel(writer, sheet_name="Stock Type Analysis", index=False)
    overall.to_excel(writer, sheet_name="Overall Summary", index=False)

print("Excel file generated successfully!")
```
